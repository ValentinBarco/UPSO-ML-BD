```
def calculate_zscore(series):
    return (series - series.mean()) / series.std()


Data1['zscore_x'] =  calculate_zscore(Data1['x'])

# M√©tricas para el grupo 1 (Data1)
metrics_data1_x = {
    'Media': Data1['x'].mean(),
    'Mediana': Data1['x'].median(),
    'Desviaci√≥n Est√°ndar': Data1['x'].std(),
    'M√≠nimo': Data1['x'].min(),
    'M√°ximo': Data1['x'].max()
}

metrics_data1_y = {
    'Media': Data1['y'].mean(),
    'Mediana': Data1['y'].median(),
    'Desviaci√≥n Est√°ndar': Data1['y'].std(),
    'M√≠nimo': Data1['y'].min(),
    'M√°ximo': Data1['y'].max()
}
import pandas as pd


df_metrics = pd.DataFrame({
    'Data1 - x': metrics_data1_x,
    'Data1 - y': metrics_data1_y
})



df_metrics

```

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()


data_uniform = np.random.uniform(low=0, high=1, size=1000)
data_normal = np.random.normal(loc=0, scale=1, size=1000)
data_binomial = np.random.binomial(n=10, p=0.1, size=1000)
data_poisson = np.random.poisson(lam=3, size=1000)
data_exponential = np.random.exponential(scale=1, size=1000)
data_chisquare = np.random.chisquare(df=2, size=1000)
data_beta = np.random.beta(a=5, b=2, size=1000)
data_gamma = np.random.gamma(shape=2, scale=1, size=1000)


fig, axs = plt.subplots(4, 2, figsize=(14, 14))


sns.kdeplot(data_uniform, ax=axs[0, 0], color='skyblue', shade=True)
axs[0, 0].set_title('Distribuci√≥n Uniforme')

sns.kdeplot(data_normal, ax=axs[0, 1], color='salmon', shade=True)
axs[0, 1].set_title('Distribuci√≥n Normal')

sns.kdeplot(data_binomial, ax=axs[1, 0], color='lightgreen', shade=True)
axs[1, 0].set_title('Distribuci√≥n Binomial')

sns.kdeplot(data_poisson, ax=axs[1, 1], color='gold', shade=True)
axs[1, 1].set_title('Distribuci√≥n Poisson')

sns.kdeplot(data_exponential, ax=axs[2, 0], color='orchid', shade=True)
axs[2, 0].set_title('Distribuci√≥n Exponencial')

sns.kdeplot(data_chisquare, ax=axs[2, 1], color='tan', shade=True)
axs[2, 1].set_title('Distribuci√≥n Chi Cuadrado')

sns.kdeplot(data_beta, ax=axs[3, 0], color='lightcoral', shade=True)
axs[3, 0].set_title('Distribuci√≥n Beta')

sns.kdeplot(data_gamma, ax=axs[3, 1], color='lightskyblue', shade=True)
axs[3, 1].set_title('Distribuci√≥n Gamma')

# Ajustar el layout
plt.tight_layout()
plt.show()

```
