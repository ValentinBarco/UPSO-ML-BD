```python
import pandas as pd

# Aquí estamos usando la función read_json de pandas para cargar el archivo JSON directamente en un DataFrame llamado datos.
datos = pd.read_json('/content/drive/MyDrive/JSON/mediciones_complejas.json')


#Creamos una lista vacía llamada dfs que almacenará DataFrames temporales.
#Estos DataFrames temporales serán generados a partir de las mediciones de cada sensor.
dfs = []
#Iteramos por cada fila del DataFrame datos usando el método iterrows(). row será una Serie con la información de la fila actual.
for _, row in datos.iterrows():
    area_nombre = row['areas']['nombre'] #Entramos al nombre
    for sensor in row['areas']['sensores']: #Entramos a sensores
        df_temp = pd.DataFrame(sensor['mediciones']) #Cargamos Mediciones en df_temp
        df_temp['Area'] = area_nombre #Cargamos el area
        df_temp['Codigo'] = sensor['codigo']
        dfs.append(df_temp)

# Concatenar todos los DataFrames temporales
df = pd.concat(dfs, ignore_index=True)
df
```

## Metodo con normalizacion

```python
# Uso de json_normalize para aplanar la estructura
df = pd.json_normalize(datos['areas'],
                       record_path=['sensores', 'mediciones'],
                       meta=[['sensores', 'codigo'], 'nombre'])

# Renombrando columnas para que sean más claras
df = df.rename(columns={
    "fecha": "Fecha",
    "temperatura": "Temperatura",
    "humedad": "Humedad",
    "presion": "Presion",
    "sensores.codigo": "Codigo",
    "nombre": "Area"
})

# Ver las primeras filas del dataframe
df
```
